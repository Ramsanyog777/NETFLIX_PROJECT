```python
### refiltering
# removing the customers who have give less rating
## removing all the movies which are having less rating
```

```python
## count th rating of every movie_id
dataset_movie_summary = df.groupby("movie_id")["Rating"].agg(["count"])
```

```python
movie_benchmark = round(dataset_movie_summary["count"].quantile(0.6), 0)
```

```python
movie_benchmark
```

```python
# if any movie has rating less than 908 I am removing it
```

```python
drop_movie_list = dataset_movie_summary[dataset_movie_summary['count'] < movie_benchmark].index
drop_movie_list
```

```python
## remove customer who have rated less number of movies
```

```python
dataset_cust_summary = df.groupby("Cust_Id")["Rating"].agg(["count"])
```

```python
cust_benchmark = round(dataset_cust_summary["count"].quantile(0.6),0)
```

```python
cust_benchmark
```

Here 36 means the customer who have given the rating less than 36 we are not going to take that for analysis

```python
drop_cust_list = dataset_cust_summary[dataset_cust_summary["count"] < cust_benchmark].index
drop_cust_list
```

```python
len(drop_cust_list)
```

```python
df = df[~df["movie_id"].isin(drop_movie_list)]
df = df[~df["Cust_Id"].isin(drop_cust_list)]
```
